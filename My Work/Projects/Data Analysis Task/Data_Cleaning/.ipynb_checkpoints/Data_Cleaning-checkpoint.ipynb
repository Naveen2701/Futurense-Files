# Problem Statement `2` : (Rename the Column names)

```python
col_re = { 'State name' : 'State/UT' ,
'District name' : 'District' ,
'Male_Literate' : 'Literate_Male' ,
'Female_Literate' : 'Literate_Female' ,
'Rural_Households' : 'Households_Rural' ,
'Urban_Households' : 'Households_Urban' ,
'Age_Group_0_29' : 'Young_and_Adult' ,
'Age_Group_30_49' : 'Middle_Aged' ,
'Age_Group_50' : 'Senior_Citizen' ,
'Age not stated' : 'Age_Not_Stated' }
```

```python
df.rename(columns = col_re, inplace=True)
```

```python
df.head()
```

# Problem Statement `3` : (Rename State/UT Names)

```python
df['State/UT'].unique()
```

```python
def format_state_name(state_name):
    words = state_name.split()
    formatted_words = [word.lower() if word.lower() == 'and' else word.capitalize() for word in words]
    return ' '.join(formatted_words)
```

```python
df['State/UT'] = df['State/UT'].apply(format_state_name)
```

```python
df['State/UT'].unique()
```

# Problem Statement `4`: (New State/UT formation)

```python
df.loc[(df['State/UT'] == "Jammu and Kashmir")]
```

```python
df.loc[(df['State/UT'] == 'Andhra Pradesh')]
```

```python
# Read Telangana districts from the text file
telangana_districts = []
with open('Telangana.txt', 'r') as file:
    telangana_districts = [line.strip() for line in file]

# Read Ladakh districts from the text file
ladakh_districts = ['Leh', 'Kargil']

# Rename State/UT names for Telangana and Ladakh districts
df.loc[df['District'].isin(telangana_districts), 'State/UT'] = 'Telangana'
df.loc[df['District'].isin(ladakh_districts), 'State/UT'] = 'Laddakh'
```

```python
df.loc[(df['State/UT'] == "Laddakh") | (df['State/UT'] == "Telangana")]
```

# Problem Statement `5` : (Find and process Missing Data)

```python
df.isnull().sum()
```

```python
missing_percentage_before = (df.isnull().sum() / len(df)) * 100
print((missing_percentage_before))
```

```python
df['Population'].fillna(df['Male'] + df['Female'], inplace=True)
```

```python
df['Male'].fillna(df['Population'] - df['Female'], inplace=True)
```

```python
df['Female'].fillna(df['Population'] - df['Male'], inplace=True)
```

```python
df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'], inplace=True)
```

```python
df['Literate_Female'].fillna(df['Literate'] - df['Literate_Male'], inplace=True)
```

```python
df['Literate_Male'].fillna(df['Literate'] - df['Literate_Female'], inplace=True)
```

```python
df['Households_Rural'].fillna(df['Households'] - df['Households_Urban'], inplace=True)
```

```python
df['Households_Urban'].fillna(df['Households'] - df['Households_Rural'], inplace=True)
```

```python
df['Households'].fillna(df['Households_Rural'] + df['Households_Urban'], inplace=True)
```

```python
df['Young_and_Adult'].fillna(df['Population'] - df['Middle_Aged'] - df['Senior_Citizen'] - df['Age_Not_Stated'], inplace=True)
```

```python
df['Middle_Aged'].fillna(df['Population'] - df['Young_and_Adult'] - df['Senior_Citizen'] - df['Age_Not_Stated'], inplace=True)
```

```python
df['Senior_Citizen'].fillna(df['Population'] - df['Young_and_Adult'] - df['Middle_Aged'] - df['Age_Not_Stated'], inplace=True)
```

```python
df['Age_Not_Stated'].fillna(df['Population'] - df['Young_and_Adult'] - df['Middle_Aged'] - df['Senior_Citizen'], inplace=True)
```

```python
missing_percentage_after = (df.isnull().sum() / len(df)) * 100
print(missing_percentage_after)
```

```python
mpb=[]
mpa=[]
for i in missing_percentage_before:
    mpb.append(i)
for i in missing_percentage_after:
    mpa.append(i)
```

```python
plt.figure(figsize=(15, 8))
X = df.columns.tolist()
Y = mpb
Z = mpa

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Y, 0.35, label='Before Cleaning')
plt.bar(X_axis + 0.2, Z, 0.35, label='After Cleaning')

plt.xticks(X_axis, X, rotation=45, ha='right')
plt.xlabel("Census Data")
plt.ylabel("Null Value")
plt.title("Census Data Cleaning")
plt.legend()
plt.tight_layout()
plt.show()
```

# Problem Statement `7`: (Process the relevant data)

```python
h_df = pd.read_csv("housing.csv")
```

```python
h_df.head()
```

```python
h_df.info()
```

```python
h_df.describe()
```

```python
req_hcol = ['District Name','Rural/Urban','Total Number of households','Total Number of Livable','Total Number of Dilapidated','Latrine_premise']
```

```python
h_df = h_df[req_hcol]
```

```python
h_df
```

```python
c_df = pd.read_csv("Census_Before_Clean.csv")
```

```python
relevant_columns = ['District Name', 'Rural/Urban', 'Total Number of households', 
                    'Total Number of Livable', 'Total Number of Dilapidated', 'Latrine_premise']

relevant_housing_data = h_df[relevant_columns]

merged_data = pd.merge(relevant_housing_data, c_df, left_on='District Name', right_on='District name', how='left')

merged_data['Households_Rural'] = merged_data['Total Number of households'] * merged_data['Rural/Urban'].apply(lambda x: 1 if x == 'Rural' else 0)
merged_data['Households_Urban'] = merged_data['Total Number of households'] * merged_data['Rural/Urban'].apply(lambda x: 1 if x == 'Urban' else 0)
merged_data['Households_Rural_Livable'] = merged_data['Rural_Households'] * merged_data['Total Number of Livable'] / 100
merged_data['Households_Rural_Dilapidated'] = merged_data['Rural_Households'] * merged_data['Total Number of Dilapidated'] / 100
merged_data['Households_Rural_Toilet_Premise'] = merged_data['Rural_Households'] * merged_data['Latrine_premise'] / 100
merged_data['Households_Urban_Livable'] = merged_data['Urban_Households'] * merged_data['Total Number of Livable'] / 100
merged_data['Households_Urban_Dilapidated'] = merged_data['Urban_Households'] * merged_data['Total Number of Dilapidated'] / 100
merged_data['Households_Urban_Toilet_Premise'] = merged_data['Urban_Households'] * merged_data['Latrine_premise'] / 100

merged_data.head()
```

```python
final_columns = ['District Name', 'Rural_Households', 'Households_Rural_Livable', 'Households_Rural_Dilapidated', 'Households_Rural_Toilet_Premise',
                 'Urban_Households', 'Households_Urban_Livable', 'Households_Urban_Dilapidated', 'Households_Urban_Toilet_Premise']

final_data = merged_data[final_columns]

column_rename = {
    'District Name': 'District',
    'Rural_Households': 'Households_Rural',
    'Households_Rural_Livable': 'Households_Rural_Livable',
    'Households_Rural_Dilapidated': 'Households_Rural_Dilapidated',
    'Households_Rural_Toilet_Premise': 'Households_Rural_Toilet_Premise',
    'Urban_Households': 'Households_Urban',
    'Households_Urban_Livable': 'Households_Urban_Livable',
    'Households_Urban_Dilapidated': 'Households_Urban_Dilapidated',
    'Households_Urban_Toilet_Premise': 'Households_Urban_Toilet_Premise'
}

final_data = final_data.rename(columns=column_rename)
final_data
```
