# Problem Statement `3` : (Rename State/UT Names)

```python
df['State/UT'].unique()
```

```python
def format_state_name(state_name):
    words = state_name.split()
    formatted_words = [word.lower() if word.lower() == 'and' else word.capitalize() for word in words]
    return ' '.join(formatted_words)
```

```python
df['State/UT'] = df['State/UT'].apply(format_state_name)
```

# Problem Statement `5` : (Find and process Missing Data)

```python
df.isnull().sum()
```

```python
missing_percentage_before = (df.isnull().sum() / len(df)) * 100
print((missing_percentage_before))
```

```python
df['Population'].fillna(df['Male'] + df['Female'], inplace=True)
```

```python
df['Male'].fillna(df['Population'] - df['Female'], inplace=True)
```

```python
df['Female'].fillna(df['Population'] - df['Male'], inplace=True)
```

```python
df['Literate'].fillna(df['Literate_Male'] + df['Literate_Female'], inplace=True)
```

```python
df['Literate_Female'].fillna(df['Literate'] - df['Literate_Male'], inplace=True)
```

```python
df['Literate_Male'].fillna(df['Literate'] - df['Literate_Female'], inplace=True)
```

```python
df['Households_Rural'].fillna(df['Households'] - df['Households_Urban'], inplace=True)
```

```python
df['Households_Urban'].fillna(df['Households'] - df['Households_Rural'], inplace=True)
```

```python
df['Households'].fillna(df['Households_Rural'] + df['Households_Urban'], inplace=True)
```

```python
df['Young_and_Adult'].fillna(df['Population'] - df['Middle_Aged'] - df['Senior_Citizen'] - df['Age_Not_Stated'], inplace=True)
```

```python
df['Middle_Aged'].fillna(df['Population'] - df['Young_and_Adult'] - df['Senior_Citizen'] - df['Age_Not_Stated'], inplace=True)
```

```python
df['Senior_Citizen'].fillna(df['Population'] - df['Young_and_Adult'] - df['Middle_Aged'] - df['Age_Not_Stated'], inplace=True)
```

```python
df['Age_Not_Stated'].fillna(df['Population'] - df['Young_and_Adult'] - df['Middle_Aged'] - df['Senior_Citizen'], inplace=True)
```

```python
missing_percentage_after = (df.isnull().sum() / len(df)) * 100
print(missing_percentage_after)
```

```python
mpb=[]
mpa=[]
for i in missing_percentage_before:
    mpb.append(i)
for i in missing_percentage_after:
    mpa.append(i)
```

```python
plt.figure(figsize=(15, 8))
X = df.columns.tolist()
Y = mpb
Z = mpa

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Y, 0.35, label='Before Cleaning')
plt.bar(X_axis + 0.2, Z, 0.35, label='After Cleaning')

plt.xticks(X_axis, X, rotation=45, ha='right')
plt.xlabel("Census Data")
plt.ylabel("Null Value")
plt.title("Census Data Cleaning")
plt.legend()
plt.tight_layout()
plt.show()
```

# Problem Statement 11: (Create a function to alter the data to create uniformity)

```python
def fix_state_names(data):
    

    state_ut_mapping = {
        'Delhi': 'Nct Of Delhi',
        'Dadra & Nagar Haveli': 'Dadra and Nagar Haveli',
        'Jammu & Kashmir':'Jammu and Kashmir',
        'Daman & Diu':'Daman and Diu',
        'Andaman & Nicobar Islands':'Andaman and Nicobar Islands',
    }

    data['State/UT'] = data['State/UT'].map(state_ut_mapping).fillna(data['State/UT'])

    return data


fixed_dataset = fix_state_names(hospital_data)

fixed_dataset.head()
```

# Problem Statement 15: (Government healthcare facility disparity)

```python
import pandas as pd

# Load the hospital data
hospital_data = pd.read_csv('C:/Users/Navee/Desktop/Jain Internship Project/DS - Python + SQL + ETL - Healthcare Census/Clean_data/government_hospital.csv')

# Load the population data (assuming you have this data)
population_data = pd.read_csv("C:/Users/Navee/Desktop/Jain Internship Project/DS - Python + SQL + ETL - Healthcare Census/Clean_data/census.csv")

# Merge hospital and population data
merged_data = pd.merge(hospital_data, population_data, on='State/UT')

# Calculate total number of beds (government + private) per population
merged_data['Total_Beds_Per_1000_People'] = (merged_data['Rural_Government_Beds'] + merged_data['Urban_Government_Beds']) / (merged_data['Population'] / 1000)

# Find the three states with the least amount of beds per population
top_states = merged_data.nsmallest(3, 'Total_Beds_Per_1000_People')

# Sort by the number of government hospitals
recommended_state = top_states.nsmallest(1, 'Rural_Government_Hospitals')

# Display the recommended state for setting up a new government hospital
print("Recommended State/UT for setting up a new government hospital:")
print(recommended_state[['State/UT', 'Rural_Government_Hospitals']])
```
